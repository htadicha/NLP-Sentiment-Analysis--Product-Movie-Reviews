```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import IPython.display as display
```

```python
df = pd.read_csv('Customers.csv')
df.head()
```

 checking the columns names and cleaning to ensure that we don't use sentence for a column names e.g  "Annual Income ($)" to "Income" and "spending score(1-100)" to "spending_score"

```python
df.columns
```

```python
df.rename(columns = {"Annual Income ($)": "Income", "Spending Score (1-100)": "Score"}, inplace= True)
df.describe()
```

```python
import plotly.graph_objects as go
import plotly.figure_factory as ff

Top=df.groupby(['Gender', 'Profession'])['Income'].sum().reset_index().sort_values(by='Income', ascending=False)
fig = ff.create_table(Top, height_constant =20)
fig.update_layout(title_text='Top Professions by Income ($)', title_x=0.3)
fig.show()


#---  showing distribution of income based on gender- for some professions ---

# fig2 = go.Figure()
# fig2.add_trace(go.Scatter(x=Top['Gender','Profession'], y=Top['Annual Income ($)'], mode='lines', name='Income Distribution'))
# fig2.update_layout(title='Income Distribution by gender for Top Professions', xaxis_title='Gender', yaxis_title='Annual Income ($)')
# fig2.show()

# fig2 = go.Figure()
# fig2.add_trace(go.Scatter(x=Top['Gender','Profession'], y=Top['Annual Income ($)'], mode='lines', name='Income Distribution'))
# fig2.update_layout(title='Income Distribution by gender for Top Professions', xaxis_title='Gender', yaxis_title='Annual Income ($)')
# fig2.show()


fig = px.line(Top, 
              x='Profession',          # X-axis will be professions
              y='Income',   # Y-axis will be income
              color='Gender',          # Separate lines by gender and color them
              title='Income Distribution by Profession, Split by Gender',
              labels={
                  "Income": "Total Income ($)",
                  "Profession": "Profession"
              })

fig.update_traces(mode='lines+markers') # Add markers to make individual professions clear
fig.update_layout(hovermode="x unified") # For a cleaner hover experience

fig.show()
```
